```python
import requests
import json
import urllib
import matplotlib.pyplot as plt
import numpy as np
from pandas import DataFrame as pd
```

```python
ip = "192.168.0.114"
port = "5000"
query = "http://{}:{}/health".format(ip, port)
r = requests.get(query)
json.loads(r.text)
```

```python
query = "http://{}:{}/getDistinct".format(ip, port)
r = requests.get(query)
inp = json.loads(r.text)
dist = pd.from_dict(inp)
symbols = dist.symbol.to_list()
```

```python
from tqdm import tqdm
import pandas

res = pandas.DataFrame()

for symbol in tqdm(symbols):
    query = "http://{}:{}/getStock/{}".format(ip, port, symbol)
    r = requests.get(query)
    inp = json.loads(r.text)
    df = pd.from_dict(inp)
    df = df.sort_values(by='date')
    if res.empty:
        res = df
    else:
        res = res.append(df)
    
res
```

```python
res.to_csv("dataset.csv")
```
